```python
# look at all of the data
# it's a lot of nested dictionaries!
print(type(data))
data
```

```python
# isolate just the comments dictionary
comments = data['results']['comments']
comments
```

```python
# look at just one of the comments
a_comment = comments[2]
a_comment
```

```python
# lowercase the comment and remove punctuation
comment_content = a_comment['commentBody']
comment_content_lower = comment_content.lower()

# without regex:
# punc = '''!()-[]{};:'"\, <>./?@#$%^&*_~'''
# for word in comment_content_lower:
#     if word in punc:
#         comment_content_lower = comment_content_lower.replace(word, "")
# print(comment_content_lower)

# easier with regex:
import re
comment_stripped = re.sub(r'[^\w\s]', '', comment_content_lower)
print(comment_stripped)
# you can use the code we wrote for the lexical richness exercise to do this
```

```python
# split text into list
comment_list = comment_stripped.split()
print(comment_list)

# create a dictionary of word frequencies!
freq_dict = {}
# go through each word in text_data
    # if the word isn't in the dictionary, add it as a key AND give it a value of 1
    # if the word IS in the dictionary, add 1 to the value
for i in comment_list:
    if i not in freq_dict.keys():
        freq_dict[i] = 1
    else:
        freq_dict[i] += 1
freq_dict
```

```python
# try and plot out a chart of the dictionary we just created!
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(20,10))
plt.bar(range(len(freq_dict)), freq_dict.values(), color='grey', align="center")
plt.xticks(range(len(freq_dict)), freq_dict.keys(), rotation=90, ha="center", fontsize=8)
plt.show()
```
