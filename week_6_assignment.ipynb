```python
import requests
import json 
import csv
from datetime import datetime 
from bs4 import BeautifulSoup
```

```python
search_template = 'https://www.indeed.com/jobs?q={}&l={}'
def url_generator(jobTitle, location):
    url = search_template.format(jobTitle, location)
    return url 
```

```python
def getQualification(link):
    print(link)
    job_qualification =[]
    responseNew = requests.get(link)
    soupObjectNew = BeautifulSoup(responseNew.text, 'html.parser')
    #job_qualification = soupObjectNew.find('div','jobsearch-jobDescriptionText')
    for p in soupObjectNew.select('div.jobsearch-jobDescriptionText p'):
        #print(p.text)
        job_qualification.append(p)
        print(p.text)
    return job_qualification
```

```python
def lookup(jobTitle, location):
    records=[]
    url = url_generator(jobTitle, location)
    response = requests.get(url)
    soupObject = BeautifulSoup(response.text, 'html.parser')
    #print(soupObject)
    postingCards = soupObject.find_all('div', 'jobsearch-SerpJobCard')
    #print(len(postingCards))

    for posting in postingCards:
        full_url = "https://www.indeed.com" + posting.find('a', 'jobtitle').get('href')
        qualification = getQualification(full_url)
        records.append(qualification)
```

```python
lookup('ux design', 'San Diego, CA')
```
